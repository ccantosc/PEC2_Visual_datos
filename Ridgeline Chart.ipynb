```python
import pandas as pd
import matplotlib.pyplot as plt
from joypy import joyplot
```

```python
datos = pd.read_excel("datos_vis.xlsx")
```

```python
plt.figure(figsize=(10, 8))
joyplot(datos, by="Comunidades y Ciudades Autonomas", column="Total", colormap=plt.cm.viridis)

plt.title("Ridgeline Chart")
plt.xlabel("INC")

plt.show()
```
